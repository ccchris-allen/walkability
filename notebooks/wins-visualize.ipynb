## Visualize the Distribution of WINS Scores

```python
names = ['92130', '92113', '92120']

for n in names:
    gdf = gpd.read_file('data/{}-street-scores.shp'.format(n))

    scores = np.clip(100 * gdf['score_rati'].values, 0, 100)
    fig, ax = plt.subplots()
    sns.distplot(scores, kde=False, ax=ax)

    plt.tick_params(axis='both', which='major', labelsize=12)

    fig.set_size_inches(12, 8)
    ax.set_xlim([0, 100])

    fig.suptitle("Distribution of WINS Scores \n {}".format(n), fontsize=20);
```

## Create Choropleth Maps for WINS Scores

```python
def plot_wins_scores(net, score_att, width=15, title=''):
    node_gdf, edge_gdf = ox.graph_to_gdfs(net.net)
    ymin, xmin, ymax, xmax = node_gdf.total_bounds

    height = width * (xmax - xmin) / (ymax - ymin)
    
    scores = [attr.get(score_att, 0) + 0.000001 for u, v, attr in net.net.edges(data=True)]

    #norm = colors.LogNorm(vmin=.00001, vmax=1.)
    norm = colors.Normalize(vmin=.0, vmax=1.)
    cmap = cm.ScalarMappable(norm=norm, cmap=cm.seismic)
    score_colors = [cmap.to_rgba(s, alpha=1.) for s in scores]

    # color the edges in the original graph with closeness centralities in the line graph
    fig, ax = ox.plot_graph(net.net, node_size=10, node_color='#ffffff', node_zorder=1, node_alpha=1,
                            edge_color=score_colors, edge_linewidth=0.5, edge_alpha=1, 
                            fig_width=width, fig_height=height, close=False, save=False, show=False)
    
    ax.set_title(title)
```

```python
plot_wins_scores(snet92130, 'score_rating', title='92130')
```

```python
plot_wins_scores(snet92120, 'score_rating', title='92120')
```
