# SPATIAL NETWORK DEMO

This notebook demonstrates the basic spatial network functionality provided by the code.  These functions are used to calculate WINS scores.

---

```python
import sys
sys.path.append("..") 

from spatial_networks import SpatialNetwork, Catchment, Path, METERS_PER_MILE
import numpy as np
import pickle

%matplotlib inline
```

```python
# north, south, east, west boundaries for zip code 92120
nsew = (32.821349, 32.773731, -117.039413, -117.104387)

net = SpatialNetwork.from_osm_bbox(nsew)
```

Plot entire network for 92120:

```python
net.plot(node_size=0, fig_height=8);
```

Simple example for calculating shortest path between two nodes.  Here we'll use calculate the shortest path within a walkability catchment area:

```python
# pick random node to do analysis on 
nodes = list(net.net.nodes())
center = nodes[100]

# get catchment area around node
catchment = net.to_catchment(center, METERS_PER_MILE)

# pick random nodes within catchment area
c_nodes = list(catchment.net.nodes())
destinations = np.random.choice(c_nodes, 10)

# calculate shortest path to all 
paths = [catchment.shortest_path(d) for d in destinations]
gdf = Path.paths_to_gdf(paths)
```

Display shortest path from center of catchment area to all other selected destination nodes in catchment:

```python
catchment.plot_paths(paths, fig_width=7);
```

Now we'll demonstrate redundant path calculation by getting redundant paths (with a redundancy coefficient under 1.10) between center and a randomly selected nearby point in catchment area:

```python
paths = catchment.redundant_paths(destinations[5], 1.10)
print "There are {} redundant paths that are 10% longer than the shortest path!".format(len(paths))
```

Plot all redundant paths found in previous step:

```python
catchment.plot_paths(paths, fig_width=7, path_alpha=0.1);
```

Determining the best and worst paths from all redundant paths.  For this simple example, we're only using distance as the criterion:

```python
best = Path.find_optimal_path(paths, 'label', criteria=['length'], minimize=[True]);
worst = Path.find_optimal_path(paths, 'label', criteria=['length'], minimize=[False])
```

Plot the *most optimal* path:

```python
catchment.plot_paths([best], fig_width=7);
```

Plot *least optimal* redundant path with a redundancy coefficient under 1.10:

```python
catchment.plot_paths([worst], fig_width=7);
```
